#### Contexto de los datos:
Una red de libros sobre política estadounidense publicados alrededor de la época de las elecciones presidenciales de 2004 y vendidos por la librería en línea Amazon.com. Los enlaces entre los libros representan la compra frecuente de libros por los mismos compradores. La red fue compilada por V. Krebs y no está publicada, pero se puede encontrar en el sitio web de Krebs. Gracias a Valdis Krebs por el permiso para publicar estos datos en este sitio web.

### ¿Que?
Lo que se puede ver en este diagrama son los 105 libros que más se vendieron en la epoca de las elecciones presidenciales en el año 2004 en Estado Unidos que estan en los nodos, representando a través de los enlaces la compra frecuente de los mismos usuarios de Amazon, siendo que los libros más populares y que tienen más conexiones entre los usuarios son The price of loyalty, American Disnaty y A National Party No More.

### ¿Por Que?
La razón de usar un diagrama de red es que es una de las formas más optimas de poder observar las 411 relaciones de frecuencia de compra de los usuarios de los 105 libros sobre politica en las elecciones del 2004, en particular se puede ver que los que tienen más conexiones son los más populares.

### ¿Cómo?
La base de datos estaba compuesta de nodos y enlaces por lo que la visualización de red era optima para este caso, en este caso no se opta por hacer que la visualización sea interactiva debido a la gran cantidad de nodos que al ser manipulados los enlaces y los demás nodos se moverian de tal forma que determinar las conexiones sería aun más dificil.

### Marca 
La marca implementada son los enlaces, ya que estos son los que unen los diferentes nodos por lo que se pueden relacionar las frecuencias de la compra de libros por los mismos usuarios.

### Canal
El canal efectuado en la visualización es de conexión ya que usando la marca de enlaces podemos a través de las lineas del canal identificar que nodo esta conectado a los demás, por lo que esta misma visualización no esta dirigida debido a que no hay una sucesión en las frecuencias

### Cumplimiento de reglas
El 2D esta justificado debido a que no se cuentan con más variables para poder ver si hay relaciones más fuertes que otras o alguna separación de los nodos por categorias. No se implementa una visualización en 3D debido a que al ser demasiados nodos el agregar el eje z, haria que para el usuario se le dificulte identificar facilmente las relaciones.
